```python
import pandas as pd
from sqlalchemy import create_engine

# Connection with the engine of the IDE
engine = create_engine('mysql+pymysql://root:password@localhost/data_exploration')

# Query to be accessed from the database server
query = "SELECT * FROM school_type_count"

# SAVED VIEW directly into Dataframe
school_types = pd.read_sql(query, con=engine)
```

## Bar Plots

**Creating a bar plot for the given categories and their counts**

```python
school_types.count()
```

```python
# Defining the different colors for the different Visualizations
colors = ['#003bcf',
          '#00eda6',
          '#e1ed00']
plt.figure(figsize=(12,8))
plt.bar(
    school_types['Types_of_school'],
    school_types['total_counts'],
    color=colors,
    edgecolor='black',
    width=0.8,
    label='Total Schools'
)
plt.title("Bar Plot for the Chicago schools according to their types",
          color='blue',
          fontsize=22)
plt.xlabel("Types of School - Chicago",color='blue')
plt.ylabel("Frequency of schools in Chicago",color='blue')
plt.legend()
```

*Hence from the above visualizations of the school datas, it is clear that the elementary schools are extremely large in number in comparision to the high schools and middle level schools*

**Visualizing the differnt attributes for the zip codes and insights for it**

```python
import pandas as pd
from sqlalchemy import create_engine


engine = create_engine('mysql+pymysql://root:password@localhost/data_exploration')

query = "SELECT * FROM zip_school_counts"

zip_counts = pd.read_sql(query, con=engine)
```

```python
zip_counts
```

```python
## Creating a bar plot using matplot properties for the given data of ZIP_Codes per School
# setting the figure size and the various properties for the histogram plot 
fig, axs = plt.subplots(2,figsize=(12,10))

axs[0].bar(
    zip_counts['ZIP_Code'].astype(str),
    zip_counts['Schools_Per_Zip'],
    color='#0ce8ce'
)
axs[0].set_xlabel("Zip Codes of State - Chicago",
                  color='#0ce8ce' )
axs[0].set_ylabel("Frequency Count of Schools - Chicago",
                  color='#0ce8ce' )
axs[0].set_title("Bar Plot for the ZIP Codes VS Count of Schools",
                 color='#0ce8ce' )

axs[1].hist(
    zip_counts['Schools_Per_Zip'],
    color='#0ce8ce',
    bins=2,
    edgecolor='black'
)
axs[0].set_xlabel("Zip Codes of State - Chicago",
                  color='#0ce8ce' )
axs[0].set_ylabel("Frequency Count of Schools - Chicago",
                  color='#0ce8ce' )
axs[0].set_title("Bar Plot for the ZIP Codes VS Count of Schools",
                 color='#0ce8ce' )
```

**Creating a scatter plot with hue value for the plot of average safety score per schools for different zip codes**

```python
engine = create_engine('mysql+pymysql://root:password@localhost/data_exploration')

query = "SELECT * FROM average_safety_zip"

average_safety_zip_df = pd.read_sql(query, con=engine)
```

```python
average_safety_zip_df.info()
```

*Selecting only the top rows for the better average and visualization of the scores with zip*

```python
average_safety_zip_df.head(5)
```

```python
top_average_safety_zip = average_safety_zip_df.head(5)
```

```python
top_average_safety_zip
```

```python
# Above is the data for the top 5 rows having most number of schools for given zip codes in that region 

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=top_average_safety_zip,
    x='average_safety_zip',
    y='number_of_schools',
    hue='ZIP_Code',
    palette='tab10',
    s=50
)
plt.xlabel('Number of Schools')
plt.ylabel('Average Safety Score')
plt.title('Safety Scores vs Number of Schools per ZIP Code')
plt.show()
```

**Hence the interpretation of the number of schools VS average safety score for the given dataset were interpereted**
